```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Video games.csv")

df['Year'].fillna(df['Year'].median(), inplace=True)
df['Publisher'].fillna('Unknown', inplace=True)

plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Global_Sales', data=df, estimator='sum', ci=None, lw=2)
plt.fill_between(df['Year'], df['Global_Sales'], color='skyblue', alpha=0.4)
plt.title('Trend of Global Sales Over the Years')
plt.xlabel('Year')
plt.ylabel('Global Sales (in millions)')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(df['Global_Sales'], bins=20, kde=False, color='skyblue')
plt.title('Distribution of Global Sales')
plt.xlabel('Global Sales (in millions)')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='NA_Sales', y='EU_Sales', data=df, alpha=0.7, color='orange')
plt.title('Relationship between North American Sales and European Sales')
plt.xlabel('North American Sales (in millions)')
plt.ylabel('European Sales (in millions)')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='JP_Sales', y='Other_Sales', data=df, alpha=0.7, color='green')
plt.title('Relationship between Japanese Sales and Other Sales')
plt.xlabel('Japanese Sales (in millions)')
plt.ylabel('Other Sales (in millions)')
plt.show()

platform_sales = df.groupby(['Year', 'Platform'])['Global_Sales'].sum().unstack().fillna(0)
platform_sales.plot.area(stacked=True, colormap='viridis', figsize=(12, 6))
plt.title('Distribution of Sales by Platform Over the Years')
plt.xlabel('Year')
plt.ylabel('Global Sales (in millions)')
plt.show()

top_publishers = df.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_publishers.index, y=top_publishers.values, palette='muted')
plt.title('Top 10 Publishers by Global Sales')
plt.xlabel('Publisher')
plt.ylabel('Global Sales (in millions)')
plt.xticks(rotation=45)
plt.show()

genre_sales = df.groupby('Genre')['Global_Sales'].sum()
plt.figure(figsize=(10, 10))
plt.pie(genre_sales, labels=genre_sales.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Distribution of Genres by Global Sales')
plt.show()
```
